```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

from rich import print
```

```python
# Load the DataFrame from the pickle file
df = pd.read_pickle("pickles/dataset.pkl")

df_filtered_ding = df[df["ding3"] > 600]

for row in df_filtered_ding.iterrows():
    print(row)

df_ding_540_545 = df[(df["ding3"] > 540) & (df["ding3"] < 545)]

print(df_ding_540_545)
```

```python
# Compute and print number of unique gamelan in the 540-545Hz ding range
gamelans_540_545 = df_ding_540_545[["Region", "Gamelan"]].drop_duplicates()
gamelan_count_540_545 = len(df_ding_540_545["Gamelan"].unique())
region_count_540_545 = len(df_ding_540_545["Region"].unique())

print(
    f"The number of unique gamelan with ding3 in the 540-545 Hz range is {gamelan_count_540_545}, distributed over {region_count_540_545} regions:"
)
print(gamelans_540_545)
```

```python
# in the datafreame, take the rows with equal region, gamelan and flavour and compute the mean of the selected columns
df_grouped = df.groupby(["Region", "Gamelan", "Flavor"]).mean()
print(df_grouped)

selisir_notes = ["ding3", "dong3", "deng3", "dung3", "dang3", "ding4"]
df_selisir_notes = df_grouped[selisir_notes]

# define a new  dictionary to store the cents values
dict_selisir_cents = {}

selisir_cents = pd.DataFrame()
for column in df_selisir_notes.columns:
    # create the key for the dictionary
    key = column + "_cents"

    # store the values in a numpy array
    values = df_selisir_notes[column].values

    # get the minimum value
    min_value = values.min()

    for value in values:
        # compute the difference in cents
        cents = 1200 * np.log2(value / min_value)

        # append the value to the value in the corresponding key
        if key in dict_selisir_cents:
            dict_selisir_cents[key].append(cents)
        else:
            dict_selisir_cents[key] = [cents]


# access each key and value in the dictionary and plot a histogram with seaborn
for key, value in dict_selisir_cents.items():
    selisir_cents[key] = value

    # plot the histogram
    sns.histplot(value, kde=True, binwidth=20)
    plt.title(key)
    plt.show()
```

```python
# get the ding3 with the highest frequency and the lowest frequency and plot them and their range and their corresponding audio
ding3_max = df["ding3"].idxmax()
ding3_min = df["ding3"].idxmin()
range_ding3 = df["ding3"].max() - df["ding3"].min()

print(
    f'The gamelan with the highest ding3 frequency is {df.iloc[ding3_max]["Gamelan"]} with a frequency of {df.iloc[ding3_max]["ding3"]} Hz'
)

print(
    f'The gamelan with the lowest ding3 frequency is {df.iloc[ding3_min]["Gamelan"]} with a frequency of {df.iloc[ding3_min]["ding3"]} Hz'
)

print(f"The range of ding3 frequencies is {range_ding3} Hz")


plt.figure(figsize=(16, 6))  # Set the size of the plot
sns.histplot(
    df["ding3"], kde=True, binwidth=1
) # Using histogram as an example, you can replace with any plot
plt.title("Histogram of ding3")  # Set the title of the plot
plt.xlabel("ding3")  # Set the x-axis label
plt.ylabel("Frequency")  # Set the y-axis label
plt.grid(True)  # Add grid lines
plt.show()  # Display the plot
```

```python
# create a dictionary to store the mean and standard deviation of the selected columns
mean_std_dict = {}
for column in df_selisir_notes.columns:
    # store the mean and standard deviation of the column in the dictionary with two decimal places
    mean_std_dict[column] = [
        round(np.mean(df_selisir_notes[column]), 2),
        round(np.std(df_selisir_notes[column]), 2),
    ]

# print the dictionary
print(mean_std_dict)


# plot a comparison of the gaussian distributions of the selected columns in one plot
plt.figure(figsize=(16, 6))
for column in df_selisir_notes.columns:
    sns.kdeplot(df[column], label=column, fill=True)
plt.title("Comparison of the Gaussian distributions of the selected columns")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Density")
plt.grid(True)
plt.show()


# plot a comparison of the mean and standard deviation of the selected columns, just these two values
plt.figure(figsize=(6, 3))
for column in df_selisir_notes.columns:
    plt.scatter(np.mean(df[column]), np.std(df[column]), label=column)
plt.title("Comparison of the mean and standard deviation of the selected columns")
plt.xlabel("Mean")
plt.ylabel("Standard deviation")
plt.legend()
plt.grid(True)
plt.show()
```
